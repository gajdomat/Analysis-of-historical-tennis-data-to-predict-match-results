````python
# Validation cell: recompute time-aware features using strictly-prior matches
# and compare against the features already present in df_pp.
import pandas as pd
import numpy as np

# Read transformed table (should match df_pp in the notebook)
df = pd.read_csv('transformed.csv', parse_dates=['tourney_date'], dtype=str)
# Keep a working copy and coerce date column
df['tourney_date'] = pd.to_datetime(df['tourney_date'].astype(str), format='%Y%m%d', errors='coerce')

# Create numeric keys for reliable sorting/merging; fill missing with -1
df['p1_key'] = pd.to_numeric(df['player_1_id'], errors='coerce').fillna(-1).astype(int)
df['p2_key'] = pd.to_numeric(df['player_2_id'], errors='coerce').fillna(-1).astype(int)

# Build long-form per-player table (use numeric player_key)
p1 = df[['tourney_date','surface','p1_key','player_1_name','player_2_id','result']].copy()
p1 = p1.rename(columns={'p1_key':'player_key','player_1_name':'player_name'})
p1['win'] = pd.to_numeric(p1['result'], errors='coerce').fillna(0).astype(int)
p2 = df[['tourney_date','surface','p2_key','player_2_name','player_1_id','result']].copy()
p2 = p2.rename(columns={'p2_key':'player_key','player_2_name':'player_name'})
p2['win'] = 1 - pd.to_numeric(p2['result'], errors='coerce').fillna(0).astype(int)

long = pd.concat([p1,p2], ignore_index=True, sort=False)
long = long.sort_values(['player_key','tourney_date']).reset_index(drop=True)

# Streaks: explicit forward pass so current match isn't counted in the streak value
def compute_streaks(g):
    g = g.sort_values('tourney_date').reset_index(drop=True)
    wins = g['win'].to_numpy(dtype=int)
    n = len(wins)
    wst = np.zeros(n, dtype=int)
    lst = np.zeros(n, dtype=int)
    w = l = 0
    for i in range(n):
        wst[i] = w
        lst[i] = l
        if wins[i] == 1:
            w += 1
            l = 0
        else:
            l += 1
            w = 0
    g['win_streak_ref'] = wst
    g['lose_streak_ref'] = lst
    # surface-wise previous matches and wins (exclude current row using shift)
    g['surface_matches_ref'] = g.groupby('surface').cumcount()
    g['surface_wins_ref'] = g.groupby('surface')['win'].transform(lambda s: s.cumsum().shift(fill_value=0))
    g['surface_wr_ref'] = (g['surface_wins_ref'] / g['surface_matches_ref']).replace([np.inf, np.nan], 0)
    return g

long = long.groupby('player_key', group_keys=False).apply(compute_streaks)
long = long.sort_values(['player_key','tourney_date']).reset_index(drop=True)

# Titles: find finals winners in the transformed table and compute titles_before strictly prior to a row's date
df['is_final'] = df.get('round', pd.Series(dtype=str)).isin(['F','Final'])
finals = df[df['is_final']].copy()
finals['winner_key'] = pd.to_numeric(finals['player_1_id'], errors='coerce').fillna(-1).astype(int)
# winner after orientation depends on result column: if result==1 player_1 won, else player_2 won
finals['winner_key'] = np.where(pd.to_numeric(finals['result'], errors='coerce')==1,
                                    pd.to_numeric(finals['player_1_id'], errors='coerce').fillna(-1).astype(int),
                                    pd.to_numeric(finals['player_2_id'], errors='coerce').fillna(-1).astype(int)
                                   )
titles = finals[['winner_key','tourney_date']].dropna().rename(columns={'winner_key':'player_key'})
titles = titles.sort_values(['player_key','tourney_date']).reset_index(drop=True)
if not titles.empty:
    titles['titles_before_at_title'] = titles.groupby('player_key').cumcount()
else:
    titles = pd.DataFrame(columns=['player_key','tourney_date','titles_before_at_title'])

# Use merge_asof to attach the most recent title row strictly before each long row (allow_exact_matches=False)
long_sorted = long.sort_values(['player_key','tourney_date']).reset_index(drop=True)
titles_sorted = titles.sort_values(['player_key','tourney_date']).reset_index(drop=True)
if not titles_sorted.empty:
    merged = pd.merge_asof(long_sorted, titles_sorted, on='tourney_date', by='player_key', direction='backward', allow_exact_matches=False)
    merged['titles_before_ref'] = merged['titles_before_at_title'].fillna(0).astype(int)
    long = merged.drop(columns=[c for c in merged.columns if c.startswith('titles_before_at_title')], errors='ignore')
else:
    long['titles_before_ref'] = 0

# Build reference features and merge back onto df on numeric keys + date
features = long[['tourney_date','player_key','win_streak_ref','lose_streak_ref','surface_wr_ref','titles_before_ref']]
p1f = features.rename(columns={'player_key':'p1_key','win_streak_ref':'p1_win_streak_ref','lose_streak_ref':'p1_lose_streak_ref','surface_wr_ref':'p1_surface_wr_ref','titles_before_ref':'p1_titles_ref'})
p2f = features.rename(columns={'player_key':'p2_key','win_streak_ref':'p2_win_streak_ref','lose_streak_ref':'p2_lose_streak_ref','surface_wr_ref':'p2_surface_wr_ref','titles_before_ref':'p2_titles_ref'})
# Merge (left) so index alignment remains with original df
df_ref = df.merge(p1f, on=['p1_key','tourney_date'], how='left').merge(p2f, on=['p2_key','tourney_date'], how='left')

# Compare existing features in df (if present) with the reference recomputed ones
comparisons = {}
cols_map = [
    ('p1_win_streak','p1_win_streak_ref'),
    ('p1_lose_streak','p1_lose_streak_ref'),
    ('p1_surface_wr','p1_surface_wr_ref'),
    ('p1_titles','p1_titles_ref'),
    ('p2_win_streak','p2_win_streak_ref'),
    ('p2_lose_streak','p2_lose_streak_ref'),
    ('p2_surface_wr','p2_surface_wr_ref'),
    ('p2_titles','p2_titles_ref'),
]
for orig, ref in cols_map:
    if orig in df_ref.columns and ref in df_ref.columns:
        # treat NaNs as a value (so NaN vs 0 is a mismatch)
        a = df_ref[orig].fillna('__NaN__')
        b = df_ref[ref].fillna('__NaN__')
        neq = (a != b)
        comparisons[orig] = int(neq.sum())
    else:
        comparisons[orig] = 'missing'

# Print concise summary
print('Validation summary (rows =', len(df_ref), ')')
for k,v in comparisons.items():
    print(f'  {k}: {v}')

# Show examples of mismatches for up to 10 rows combined across features
mismatch_idx = set()
for orig, ref in cols_map:
    if orig in df_ref.columns and ref in df_ref.columns:
        a = df_ref[orig].fillna('__NaN__')
        b = df_ref[ref].fillna('__NaN__')
        neq_idx = df_ref.index[a != b].tolist()
        for ii in neq_idx:
            mismatch_idx.add(ii)
        if len(mismatch_idx) >= 10:
            break
mismatch_idx = sorted(list(mismatch_idx))[:10]
if mismatch_idx:
    print('Examples (up to 10) of rows with differences:')
    display(df_ref.loc[mismatch_idx, [
        'tourney_date','player_1_id','player_2_id',
        'p1_win_streak','p1_win_streak_ref','p1_surface_wr','p1_surface_wr_ref','p1_titles','p1_titles_ref',
        'p2_win_streak','p2_win_streak_ref','p2_surface_wr','p2_surface_wr_ref','p2_titles','p2_titles_ref'
    ]])
else:
    print('No mismatches found on the sampled comparisons.')
df_final=df_final.drop_duplicates()
df_final.shape
````

````python
df=pd.read_csv("procesed.csv")

date_str = df["tourney_date"].astype(str)

bad_tourneys = (
    df.groupby("tourney_name")
      .agg(
          tourney_ids=("tourney_id", "nunique"),
          surfaces=("surface", "nunique"),
          draw_sizes=("draw_size", "nunique"),
          levels=("tourney_level", "nunique"),
      )
      .query("tourney_ids > 1 or surfaces > 1 or draw_sizes > 1 or levels > 1")
)

df["tourney_name"] = (
    df["tourney_name"]
    .str.replace(r"\s+\d+$", "", regex=True)
)

parsed = pd.to_datetime(
    date_str,
    format="%Y%m%d",
    errors="coerce"
)

calendar_mask = parsed.notna()
bad_calendar = df[~calendar_mask]


string_cols = df.select_dtypes(include="object").columns
df[string_cols] = (
    df[string_cols]
    .apply(lambda s: s.str.replace(r"\s+", " ", regex=True).str.strip())
)

violations = (
    df.groupby("tourney_id")["match_num"]
      .nunique()
      .ne(df.groupby("tourney_id")["match_num"].size())
)

bad_tourneys = violations[violations].index

df_bad = df[df["tourney_id"].isin(bad_tourneys)] \
           .sort_values(["tourney_id", "match_num"])

print(df_bad[["tourney_id", "match_num"]].to_string(index=False))


players = pd.concat([
    df[["winner_id", "winner_name"]]
        .rename(columns={"winner_id": "player_id", "winner_name": "player_name"}),
    df[["loser_id", "loser_name"]]
        .rename(columns={"loser_id": "player_id", "loser_name": "player_name"}),
])

players = players[players["player_name"] == "Martin Landaluce"]

players["player_id"].value_counts()

ID_MAP = {
    212021: 211776
}
df["winner_id"] = df["winner_id"].replace(ID_MAP)
df["loser_id"] = df["loser_id"].replace(ID_MAP)

df["winner_entry"] = df["winner_entry"].str.upper()

df["loser_entry"] = df["loser_entry"].str.upper()

df["winner_name"] = df["winner_name"].str.replace(r"\s*Gb42$", "", regex=True)
df["loser_name"] = df["loser_name"].str.replace(r"\s*Gb42$", "", regex=True)

players = pd.concat([
    df[["winner_id", "winner_hand"]].rename(
        columns={"winner_id": "player_id", "winner_hand": "player_hand"}
    ),
    df[["loser_id", "loser_hand"]].rename(
        columns={"loser_id": "player_id", "loser_hand": "player_hand"}
    ),
])
players["player_hand"] = players["player_hand"].str.upper().str.strip()



def canonical_hand(sub):
    counts = sub[sub != 'U'].value_counts()
    if not counts.empty:
        return counts.idxmax()
    return 'U'  # if all unknown

canonical_hands = players.groupby("player_id")["player_hand"].apply(canonical_hand)


hand_map = canonical_hands.to_dict()

df["winner_hand"] = df["winner_id"].map(hand_map)
df["loser_hand"] = df["loser_id"].map(hand_map)

df["winner_age"] = df["winner_age"].round().astype("Int64")
df["loser_age"] = df["loser_age"].round().astype("Int64")
df = df.drop(["winner_seed", "winner_entry", "loser_seed", "loser_entry"], axis=1)

import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer

obj_cols = df.select_dtypes(include="object").columns

category_maps = {}

for col in obj_cols:
    df[col] = df[col].astype("category")
    category_maps[col] = df[col].cat.categories
    df[col] = df[col].cat.codes.replace(-1, np.nan)

imputer = KNNImputer(n_neighbors=5, weights="distance")
df_imputed = pd.DataFrame(
    imputer.fit_transform(df),
    columns=df.columns,
    index=df.index
)

for col in obj_cols:
    df_imputed[col] = (
        df_imputed[col]
        .round()
        .astype(int)
        .clip(0, len(category_maps[col]) - 1)
        .map(dict(enumerate(category_maps[col])))
    )

assert not df_imputed[obj_cols].isna().any().any(), "Missing values remain in object columns"

df = df_imputed
df.head()
````

````python
import pandas as pd

# Columns that are match-level (same for both players)
match_cols = [
    "tourney_id", "tourney_name", "surface", "draw_size",
    "tourney_level", "tourney_date", "match_num",
    "score", "best_of", "round", "minutes"
]

# Winner → player_1 mapping
winner_map = {
    "winner_id": "player_1_id",
    "winner_name": "player_1_name",
    "winner_hand": "player_1_hand",
    "winner_ht": "player_1_ht",
    "winner_ioc": "player_1_ioc",
    "winner_age": "player_1_age",
    "winner_rank": "player_1_rank",
    "winner_rank_points": "player_1_rank_points",
    "w_ace": "player_1_ace",
    "w_df": "player_1_df",
    "w_svpt": "player_1_svpt",
    "w_1stIn": "player_1_1stIn",
    "w_1stWon": "player_1_1stWon",
    "w_2ndWon": "player_1_2ndWon",
    "w_SvGms": "player_1_SvGms",
    "w_bpSaved": "player_1_bpSaved",
    "w_bpFaced": "player_1_bpFaced",
}

# Loser → player_2 mapping
loser_map = {
    "loser_id": "player_2_id",
    "loser_name": "player_2_name",
    "loser_hand": "player_2_hand",
    "loser_ht": "player_2_ht",
    "loser_ioc": "player_2_ioc",
    "loser_age": "player_2_age",
    "loser_rank": "player_2_rank",
    "loser_rank_points": "player_2_rank_points",
    "l_ace": "player_2_ace",
    "l_df": "player_2_df",
    "l_svpt": "player_2_svpt",
    "l_1stIn": "player_2_1stIn",
    "l_1stWon": "player_2_1stWon",
    "l_2ndWon": "player_2_2ndWon",
    "l_SvGms": "player_2_SvGms",
    "l_bpSaved": "player_2_bpSaved",
    "l_bpFaced": "player_2_bpFaced",
}

# Build final dataframe
df_pp = pd.concat(
    [
        df[match_cols],
        df[list(winner_map)].rename(columns=winner_map),
        df[list(loser_map)].rename(columns=loser_map),
    ],
    axis=1
)

# Binary target: player_1 always wins
df_pp["result"] = 1


import numpy as np

swap_mask = np.random.rand(len(df_pp)) < 0.5

p1_cols = [c for c in df_pp.columns if c.startswith("player_1")]
p2_cols = [c for c in df_pp.columns if c.startswith("player_2")]

df_pp.loc[swap_mask, p1_cols + p2_cols] = (
    df_pp.loc[swap_mask, p2_cols + p1_cols].values
)

df_pp.loc[swap_mask, "result"] = 1 - df_pp.loc[swap_mask, "result"]
````

````python
obj_cols = df.select_dtypes(include=['object']).columns.tolist()
for c in obj_cols:
    try:
        df[c] = df[c].astype('category').cat.codes
    except Exception:
        pass

float_cols = df.select_dtypes(include=['float']).columns.tolist()
for c in float_cols:
    s = pd.to_numeric(df[c], errors='coerce')
    non_na = s.dropna()
    if non_na.empty:
        df[c] = s.astype('Int64')
        continue
    if (non_na % 1 == 0).all():
        df[c] = s.round(0).astype('Int64')
    else:
        df[c] = s.astype(float)

for col in ['p1_win_streak','p2_win_streak','p1_lose_streak','p2_lose_streak','p1_titles','p2_titles']:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0).astype('Int64')

for col in ['p1_surface_wr','p2_surface_wr']:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0.0).astype(float)
````
